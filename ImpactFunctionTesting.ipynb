```python
import numpy as np
import ImpactFunctions
import matplotlib.pyplot as plt

times = np.linspace(0, 7, 200)

risk_averse_process = ImpactFunctions.TradingProcess(10, 1, 4, 10)
risk_averse_process.compute_volumes_and_velocities(times)

less_risk_averse_process = ImpactFunctions.TradingProcess(10, 1, 4, 2)
less_risk_averse_process.compute_volumes_and_velocities(times)

even_less_risk_averse_process = ImpactFunctions.TradingProcess(10, 1, 4, 0.1)
even_less_risk_averse_process.compute_volumes_and_velocities(times)

twap_process = ImpactFunctions.TradingProcess(10, 1, 4, 0)
twap_process.compute_volumes_and_velocities(times)



risk_averse_volumes = risk_averse_process.volume
less_risk_averse_volumes = less_risk_averse_process.volume
even_less_risk_averse_volumes = even_less_risk_averse_process.volume
twap_volumes = twap_process.volume


plt.plot(times, risk_averse_volumes)
plt.plot(times, less_risk_averse_volumes)
plt.plot(times, even_less_risk_averse_volumes)
plt.plot(times, twap_volumes)
```

```python
plt.plot(times, less_risk_averse_process.volume)
plt.plot(times, less_risk_averse_process.velocity)
```

```python
plt.plot(times, twap_process.volume)
plt.plot(times, twap_process.velocity)
```

```python
impact = ImpactFunctions.ImpactProcess(3, 0.5, 0)

twap_impact = impact.compute_impact(twap_process)

plt.plot(times, -twap_impact)

```
