```python
import numpy as np

# Temporary impact is only incurred for each trade... maybe can ignore for now
# Propagator is more long term, transient impact

class ImpactModel:
    def __init__(self, transient_impact_propagator_kernel, temporary_impact_function):
        self.propagator = transient_impact_propagator_kernel
        self.temporary_impact = temporary_impact_function

class ObishaevaWangModel(ImpactModel):
    def __init__(self, scale, decay, temporary_impact_function):
        self.scale = scale
        self.decay = decay

        def OW_propagator(s):
            return self.scale * np.exp(-self.decay * s)

        ImpactModel.__init__(self, OW_propagator, temporary_impact_function)

class ExponentialDecayPermanentImpactModel(ImpactModel):
    def __init__(self, scale , decay, permanent, temporary_impact_function):
        self.scale = scale
        self.decay = decay
        self.permanent = permanent

        def propagator(s):
            return self.scale * np.exp(-self.decay * s) + self.permanent

        ImpactModel.__init__(self, propagator, temporary_impact_function)

class PowerLawDecayPermanentImpactModel(ImpactModel):
    def __init__(self, scale, decay, power, permanent, temporary_impact_function):
        self.scale = scale
        self.decay = decay
        self.permanent = permanent
        self.power = power

        def propagator(s):
            return self.scale * np.power(self.decay * s, self.power) + self.permanent
        
        ImpactModel.__init__(self, propagator, temporary_impact_function)

class OrderBook:
    def __init__(self, bid_ask_spread):
        self.spread = bid_ask_spread

class Inventory:
    def __init__(self, inventory_volume, cash_position):
        self.inventory = inventory_volume
        self.cash = cash_position
```
